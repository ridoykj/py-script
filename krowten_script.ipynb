```python
import os
from google.cloud import firestore

google_project_id = os.getenv("GOOGLE_PROJECT_ID")
firestore_database = os.getenv("FIRESTORE_DATABASE")

if google_project_id is None or firestore_database is None:
    raise ValueError("Missing GCP_PROJECT_ID or FIRESTORE_DATABASE in env")

db = firestore.Client(project=google_project_id, database=firestore_database)
```

```python
video_assets = db.collection('video-assets')
```

```python
from google.api_core import exceptions
from google.cloud.firestore_v1.base_query import FieldFilter

try:
    channel_id = "UC4AllprYMFCXhx5OODDUX8A"
    videos = ["gS13ZlpHE28", "gS13ZlpHE24"]
    fs_path = f"channels.`{channel_id}`.`yt-video-id`"
    # query = video_assets.where(filter=FieldFilter(fs_path, "in", videos))
    query = video_assets.where(filter=FieldFilter(fs_path, "in", videos)).select([fs_path])
    docs1 = query.stream()
    for doc in docs1:
        print(f"Document ID: {doc.id}, the video is: {doc.get(fs_path)}")
    docs2 = query.stream()
    match_videos = [doc.get(fs_path) for doc in docs2]
    for vid_id in match_videos:
        print(f"Document ID: {vid_id}")
        # doc = db.collection('video-assets').document(id)
except exceptions.GoogleAPICallError as e:
    print(f"Firestore Error: {e.message}")
```

```python
items_col = db.collection("youtube-cache").document('UC-NEgl3wPCGhwh_Y8rlA7lA').collection('videos')
videos = ['3-9Cdpd9s6o', '0N1hOWOmSY4']


def chunker(seq, size):
    for i in range(0, len(seq), size):
        yield seq[i:i + size]


for chunk in chunker(videos, 499):
    refs = [items_col.document(video_id) for video_id in chunk]
    snaps = list(db.get_all(refs))
    for snap in snaps:
        print(f"data: {snap.to_dict()}")
```
